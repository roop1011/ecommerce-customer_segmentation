```python

x=df.iloc[:,2:].values
x
```

```python
#Its a preprocessing technique used to scale features to a specific rangew (usually [0,1])
scale=MinMaxScaler()
features=scale.fit_transform(x)
features
```

Elbow method to get the optimal K value

```python
inertia=[]
for i in range(1,16):
  k_means=KMeans(n_clusters=i)
  k_means=k_means.fit(features)
  inertia.append(k_means.inertia_)
#It does the elbow method to genertae the suitable clusters by using inertia method that gives the sum of the square distancews of each point with the respective assigned cluster of the point
```

```python
#Elbow graph
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.plot(range(1,16),inertia,'bo-')
plt.xlabel('No of clusters'),plt.ylabel('Inertia')
```

```python

#Kelbow visualizer
plt.subplot(1,2,2)
kmeans=KMeans()
visualize=KElbowVisualizer(kmeans,k=(1,16))
visualize.fit(features)
plt.suptitle('Elbow Graph and Elbow Visualizer')
visualize.poof()
plt.show()
#distortion refers to a metric that represents how well the data points are grouped into their assigned clusters.
```

Silhoutte Score for each k value

```python
silhouette_avg=[]
for i in range(2,16):
  #intialize kmeans cluster
  kmeans=KMeans(n_clusters=i)
  cluster_labels=kmeans.fit_predict(features)

  #Silhouette score
  silhouette_avg.append(silhouette_score(features,cluster_labels))
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(2,16),silhouette_avg,'bX-')
plt.xlabel('Values of K')
plt.ylabel('Silhouette score')

plt.title('Silhouette analysis for optimal K')
plt.show()
```

Kmeans Model Here we will take K value as 3 as per Elbow Method

```python
model=KMeans(n_clusters=3)
model=model.fit(features)
```

```python
y_km=model.predict(features)
centers=model.cluster_centers_
```

```python
df['Cluster']=pd.DataFrame(y_km)
df.to_csv("Cluster_data",index=False)
```
